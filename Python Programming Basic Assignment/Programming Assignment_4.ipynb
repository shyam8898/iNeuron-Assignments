1) Write a Python Program to Find the Factorial of a Number?

```python
def factorial(num):
    factorial = 1
    if num < 0:
        print("Sorry, factorial does not exist for negative numbers")
    elif num == 0:
        print("The factorial of 0 is 1")
    else:
        for i in range(1,num + 1):
            factorial = factorial*i
        print("The factorial of",num,"is",factorial)

# Calling Function
factorial(5)
```

2) Write a Python Program to Display the multiplication Table?

```python
def multiplication_table(num):
    for i in range(1,11):
        print(f"{num} x {i} = {num*i}")
        
# Calling Function
multiplication_table(6)
```

3) Write a Python Program to Print the Fibonacci sequence?

```python
def fibonacci_sequence(nterms):
    # first two terms
    n1, n2 = 0,1
    count = 0
    
    # check if the number of terms is valid
    if nterms <= 0:
        print("Please enter a positive integer")
    # if there is only one term, return n1
    elif nterms == 1:
        print("Fibonacci sequence upto",nterms,":")
        print(n1)
    # generate fibonacci sequence
    else:
        print("Fibonacci sequence:")
        while count < nterms:
            print(n1)
            nth = n1 + n2
            # update values
            n1 = n2
            n2 = nth
            count += 1
            
# Calling Function
fibonacci_sequence(9)
```

4) Write a Python Program to Check Armstrong Number?

```python
def armstrong_number(num):
    order = len(str(num))
    # initialize sum
    sum = 0
    # find the sum of the cube of each digit
    temp = num
    while temp > 0:
        digit = temp % 10
        sum += digit ** order
        temp //= 10
    # display the result
    if num == sum:
        print(num,"is an Armstrong number")
    else:
        print(num,"is not an Armstrong number")

# Calling Function
armstrong_number(407)
```

5) Write a Python Program to Find Armstrong Number in an Interval?

```python
def armstrong_number_interval(lst):
    for i in lst:
        # Calling above function
        armstrong_number(i)

# Calling Function
armstrong_number_interval([i for i in range(1000)])
```

6) Write a Python Program to Find the Sum of Natural Numbers?

```python
def Sum_natural_numbers(num):
    if num <= 0:
        print("Enter a positive number")
    else:
        sum=0
        # use while loop to iterate until zero
        while(num > 0):
            sum += num
            num -= 1
        print("The sum is", sum)

# Calling Function
Sum_natural_numbers(14)
```
